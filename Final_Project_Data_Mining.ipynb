# **Function**

````python
# Clean Data
def cleanText(txt):
  # Lower Case
  docx = unidecode(unescape(txt.lower().strip()))
  
  # Remove Links
  urlPattern = re.compile(r'http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\(\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+')
  docx = re.sub(urlPattern,' ', docx)
  
  # Remove Symbol, keep "_"
  docx = re.sub(r'[^\w]', ' ', docx)
  
  # Tokenisasi
  Tokens = TextBlob(docx).words 
    
  # Stopwords Filtering
  docx = ' '.join([t for t in Tokens if str(t) not in Sastrawi_StopWords_id ])
  docx = stemmer.stem(docx)
  
  return docx
````

# **Preprocessing**

## **Clean Tweet Data**

````python
# Get Tweet Column
data_tweet = datai[['tweet']]
````

````python
# Loading Stopwords
Sastrawi_StopWords_id = factory.get_stop_words()
Sastrawi_StopWords_id = set(Sastrawi_StopWords_id)

# Add Stopwords
Sastrawi_StopWords_id.add('ð')
Sastrawi_StopWords_id.add('�')
Sastrawi_StopWords_id.add('¥')
````

````python
 # Initialize New Column
data_tweet['clean_tweet'] = ''

# Clean Text
for idx, post in tqdm(data_tweet.iterrows()):
  data_tweet.at[idx, 'clean_tweet'] = cleanText(post.tweet)
````

````python
# Copy Data
data_tweet_slang = data_tweet['clean_tweet']
````

````python
# Clean Slang
slangFixId = tau.loadCorpus(file = 'data/slang.txt', sep=':')
cleanData = []
for t in tqdm(data_tweet_slang):
    doc = ' '.join(re.sub(r"(#[A-Za-z0-9]+)|(@\w+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)"," ",t).split())
    cleanDoc = tau.cleanText(doc, fix=slangFixId, lan = 'id')
    cleanData.append(cleanDoc)
````

````python
# Tranform to Data Frame
tweet_clean_data = pd.DataFrame(cleanData)
tweet_clean_data.columns = ['clean_tweet']
````

````python
# Save Data to Local Storage
tweet_clean_data.to_csv("data_clean_tweet.csv")
````

# **Centrality Analysis**

````python
# Convert Data to List
users = final_data.username.to_list()
tweets = final_data.tweet.to_list()
````

````python
# Intialize New Graph
G = nx.Graph()
G.add_nodes_from(users)

print('Banyak Vertex = ', G.number_of_nodes())
print('Banyak Edges = ', G.number_of_edges())
````

````python
# Get Tweet Data that Contains Someone's Tag
for usr, t in zip(users, tweets):
    mentionS =  re.findall("@([a-zA-Z0-9]{1,15})", t)
    for mention in mentionS:
        if "." not in mention:
            G.add_edge(usr, mention)
            
print('Banyak Vertex = ', G.number_of_nodes())
print('Banyak Edges = ', G.number_of_edges())
````

````python
# Export GexF File
qry = "final piala dunia"
nx.write_gexf(G, qry+".gexf")
````
